```python
import os
import requests

def get_weather_2(city, days) -> str:
    base_url = "https://api.weatherapi.com/v1/forecast.json"
    
    params = {
        "key": os.getenv('WEATHER_API_KEY', None),
        "q": city,
        "days": days,
        "day_fields": "maxtemp_c,mintemp_c,condition",
    }
    
    response = requests.get(base_url, params=params)
    data = response.json()
    print(data)
    forecast_days = data['forecast']['forecastday']
    output_text = []
    for day in forecast_days:
        date = day['date']
        weather_text = day['day']['condition']['text']
        max_temp_c = day['day']['maxtemp_c']
        min_temp_c = day['day']['mintemp_c']
        output_text.append(f"Date: {date}\nWeather: {weather_text}\nTemperature: {min_temp_c}°C ~ {max_temp_c}°C")
    return "\n\n".join(output_text)

get_weather_2('Beijing', '5')
```

```python
from resources import openai
import json

def embedding(text):
    return openai.embeddings.create(input=text, model='text-embedding-3-small').data[0].embedding

data = []
with open('tone.jsonl', encoding='utf-8') as tone_file:
    for line in tone_file:
        d = json.loads(line)
        d['user_embedding'] = embedding(d['user'])
        d['bad_reply_embedding'] = embedding(d['bad_reply'])
        data.append(d)

with open('tone.jsonl', 'w', encoding='utf-8') as tone_file:
    for d in data:
        tone_file.write(json.dumps(d, ensure_ascii=False) + '\n')

```
